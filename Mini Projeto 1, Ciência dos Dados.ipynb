Os métodos de distribuição teórica serão os de Poisson e binomial. A escolha do método de poisson se deu pois este método é característico por modelar dados de contagem, bem como quantidade de gás consumido por residência.
A escolha do método binomial se deu, pois a partir da análise do histograma, a imensa maioria dos locais analisados possuem consumo inferior a  2300 de consumo de gás natural. Portanto, podemos analisar qual a proporção de consumos maiores que 2300 ou menores que este valor.
