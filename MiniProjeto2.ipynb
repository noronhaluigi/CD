1.Começaremos plotando um conjunto de valores hipotéticos e aleatórios, de 0 até 10 com 100 iterações, e plotaremos um histograma, alem de calcular a esperança.

```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados1 = []
for i in range (100):
    dados1.append(random.randint(0,10))
plt.hist(dados1, normed=1)
plt.title("Histograma dos dados em 100 iterações")
plt.xlabel("Valor")
plt.ylabel("Frequência")
print(np.mean(dados1))
```

2.Separando eventos dos dados, com 3 eventos em cada um inicialmente, e calcularemos a média e desvio padrão desses valores. Logo, após 1000 iterações, plotaremos um histograma das médias obtidas desses valores, e observaremos os resultados, comparando com uma distribuição normal.

```python
medias = []
for i in range(1000):
    mean = 0
    soma = 0
    x = random.sample(dados1, 3)
    for k in range(3):
        soma += x[k]
    mean = soma/3
    medias.append(mean)

a = np.mean(medias)
b = np.std(medias)    
#s = np.random.normal(loc=a, scale=b, size=1000)
plt.hist(medias, normed=1)

#plot3=plt.hist(s, 30, normed=True, color = 'white')
#count, bins, ignored = plot3
#plt.plot(bins, 1/(b * np.sqrt(2 * np.pi)) *np.exp( - (bins - a)**2 / (2 * b**2) ),linewidth=2, color='r')

count, bins, ignored = plt.hist(medias, normed=1)
plt.plot(bins, 1/(b * np.sqrt(2 * np.pi)) *np.exp( - (bins - a)**2 / (2 * b**2) ),linewidth=2, color='r')

plt.title("Histograma das médias dos eventos de 3 eventos")
plt.xlabel("Valor")
plt.ylabel("Frequência")
print(np.mean(medias))
print(np.std(medias))
```

Podemos percebe que as médias de ambos os dados são semelhantes. Alem disso, aparentemente o histograma das médias dos eventos apresenta a característica de uma normal.

3.Vamos ampliar o tamanho da amostra, que passará a ser de 8 valores e estudar o resultado.

```python
medias2 = []
for i in range(1000):
    mean = 0
    soma = 0
    x = random.sample(dados1, 8)
    for k in range(8):
        soma += x[k]
    mean = soma/8
    medias2.append(mean)
plt.hist(medias2, normed=1)

count, bins, ignored = plt.hist(medias2, normed=1)
plt.plot(bins, 1/(b * np.sqrt(2 * np.pi)) *np.exp( - (bins - a)**2 / (2 * b**2) ),linewidth=2, color='r')

plt.title("Histograma das médias dos eventos de 8 eventos")
plt.xlabel("Valor")
plt.ylabel("Frequência")
print(np.mean(medias2))
print(np.std(medias2))
```

Vemos que o histograma se assemelha ainda mais com uma distribuição normal, e que a sua média se tornou mais semelhante com a média dos dados iniciais.

4.Como finalização, vamos tornar os dados exorbitantes para tentar aproximar cada vez mais da distribuição normal. Alteraremos para 100000 o número de itens dos dados, com uma variação de 0-100. Para a plotagem do histograma, utilizaremos 10000 eventos possuindo 50 eventos.

```python
dados2 = []
for i in range (100000):
    dados2.append(random.randint(0,100))
plt.hist(dados2, normed=1)
plt.title("Histograma dos dados em 100 iterações")
plt.xlabel("Valor")
plt.ylabel("Frequência")
print(np.mean(dados2))
print(np.std(dados2))
```

```python
medias3 = []
for i in range(10000):
    mean = 0
    soma = 0
    x = random.sample(dados2, 50)
    for k in range(50):
        soma += x[k]
    mean = soma/50
    medias3.append(mean)
plt.hist(medias3, normed=1)
a2 = np.mean(medias3)
b2 = np.std(medias3)
count, bins, ignored = plt.hist(medias3, normed=1)
plt.plot(bins, 1/(b2 * np.sqrt(2 * np.pi)) *np.exp( - (bins - a2)**2 / (2 * b2**2) ),linewidth=2, color='r')

plt.title("Histograma das médias dos eventos de 3 eventos")
plt.xlabel("Valor")
plt.ylabel("Frequência")
print(np.mean(medias3))
print(np.std(medias3))
```

CONCLUSÃO FINAL: Após todos esses testes, podemos observar a relação entre um conjunto de dados aleatórios qualquer e seus eventos com uma distribuição normal. Ao longo dos procedimentos realizados alteramos as bases de dados cada vez que o programa era rodado, e as médias permaneciam ainda semelhantes, assim como a o histograma e a curva normal. Percebe-se também que conforme aumentamos as iterações, maiores serão as semelhanças do histograma com a distribuição normal, lembrando que caso o tamanho do evento seja 1, a média é o próprio número e portanto nunca haverá uma distribuição normal (a distribuição da média será idêntica aos dados). Dessa forma, conclui-se que, tendendo as iterações ao infinito, teremos uma distribuição normal a partir das médias de eventos de um conjunto de dados de uma variável aleatória. Olhando para o desvio padrão, percebemos que conforme o tamanho da amostra cresce, menor é o desvio padrão, tornando a curva normal mais 'pontuda'.
