### Itens 4 a 6 - Resolução:

```python
import numpy as np
faixa1 = np.arange(0, 46000, 2300)
plot1 = dados.ConsumptionGJ.plot.hist(bins= faixa1, normed = 1)
plt.title("Frequência de consumo")
```

Os métodos de distribuição teórica serão os de exponencial e log normal. A escolha pela exponencial, assim como a log normal, se deu com o formato do histograma plotado. Com a função exponencial, podemos evidenciar uma situação em que temos um grande valor em números menores de x, e uma queda brusca curvilíneo conforme o x aumenta. Com a log normal, podemos perceber um aumento brusco nos primeiros valores de x, que vai decrescendo conforme x aumenta. Confirmando, assim, a semelhança com o gráfico histograma dos dados considerados.

```python
from scipy import stats
import numpy as np    
```

Analisando a função exponencial.

```python
x = sorted(dados.ConsumptionGJ)
plt.plot(x, stats.expon.pdf(x,loc=0 , scale=453385.23), label = 'expon pdf')
plt.title('frequência por quantidade de gás consumido')
```

```python
x = sorted(dados.ConsumptionGJ)
plt.plot(x, stats.expon.pdf(x,loc=0 , scale=4533.8523), label = 'expon pdf')
plt.hist(dados.ConsumptionGJ,bins= faixa1, normed = 1)
plt.ylim((0,0.00015))
plt.xlim((0,46000))
plt.title('frequência por quantidade de gás consumido')
```

Pensando na distribuição lognormal.

Estimando valores, ajustando a média e o desvio padrão.

```python
#tentando lognorm
#lognorm.pdf(x, s, loc, scale)
import math
mu = 29.01661

x = sorted(dados.ConsumptionGJ)
plt.plot(x, stats.lognorm.pdf(x, 45.338523, loc=0 , scale = math.exp(mu)), label = 'expon pdf')
plt.hist(dados.ConsumptionGJ,bins= faixa1, normed = 1)
plt.ylim((0,0.00015))
plt.xlim((0,46000))
plt.title('frequência por quantidade de gás consumido')
```

Atribuindo a função fit para estimar melhores valores para a base de dados.

```python
b = stats.lognorm.fit(dados.ConsumptionGJ, floc=0)
b
```

```python
#tentando lognorm
#lognorm.pdf(x, s, loc, scale)
import math
mu = 29.01661

x = sorted(dados.ConsumptionGJ)
plt.plot(x, stats.lognorm.pdf(x,2.5423 , loc=0 , scale = 59321.3), label = 'expon pdf')
plt.hist(dados.ConsumptionGJ,bins= faixa1, normed = 1)
plt.ylim((0,0.00015))
plt.xlim((0,46000))
plt.title('frequência por quantidade de gás consumido')
```

Conclusão: Para a distriubição exponencial, podemos ver que seu melhor aproveitamento se dá nos valores menores de x, isto é, quando a função possui maiores valores em y. Assim, consideraremos os parâmetros obtidos pela função fit, pois conseguem extrair o melhor desse aproveitamento da função. Para a distribuição lognormal, o seu melhor aproveitamento se dá com maiores valores de x. Dessa forma, consideraremos os parâmetros obtidos pela função fit, pois esses números respondem melhor a curva da base de dados no histograma.

```python
plt.plot(x, stats.lognorm.pdf(x,2.5423 , loc=0 , scale = 59321.3), label = 'lognorm pdf')
plt.hist(dados.ConsumptionGJ,bins= faixa1, normed = 1)
plt.plot(x, stats.expon.pdf(x,loc=0 , scale=2904.5512), label = 'expon pdf')
plt.ylim((0,0.00015))
plt.xlim((0,46000))
plt.title('frequência por quantidade de gás consumido')
```

Podemos perceber que ambas as funções escolhidas possuem suas particularidades, diferenças e semelhanças com a base de dados no histograma. Dessa forma, não podemos tirar conclusões precipitadas sobre a melhor função para representar os dados. Serão necessários testes de aderência.

Utilizando QQ plot.

Utilizando a técnica de QQplot do scipy, para a exponencial e lognormal, calculando erro quadrático.

```python
import numpy as np 
import pylab 
import scipy.stats as stats

#measurements = np.random.normal(loc = 20, scale = 5, size=100)   
stats.probplot(x,(0, 2904.5512), dist="expon", plot=pylab)
pylab.show()
stats.probplot(x,(2.54,0, 59321.34), dist="lognorm", plot=pylab)
pylab.show()
```

Após plotar os gráficos QQplot da exponencial e da lognormal, podemos perceber que a precisão da função é maior do que a da lognormal (0.935 de r^2 da exponencial, contra 0.479 de r^2 da lognormal).

Testando a aderência por gráficos da frequência relativa acumulada (a partir dos dados) vs a função
de distribuição acumulada.

```python
# fra e FDA
fra = [] # frequencia relativa acumulada
for i in range(len(x)):
    acc = (i - 0.5) / len(x)
    if acc < 0:
        acc = 0
    fra.append(acc)



cdf_exponencial = stats.expon.cdf(x, scale=2904.5512 ) #FDA
cdf_lognormal =  stats.lognorm.cdf(x,2.5423 , loc=0 , scale = 59321.3)


fig = plt.figure(figsize=(12,8))

plt.title("Cumulativa: Dados x Exponencial x Lognormal")

#dados_fra_plot, = plt.plot(x, fra, '-', color='black', label="F.R.A dos dados")
cdf_exp_plot,   = plt.plot(x, cdf_exponencial, '-', color='red', label = "Cumulativa da exponencial")
cdf_lognormal_plot, = plt.plot(x, cdf_lognormal, '-', color='green', label = "Cumulativa da distribuição Lognormal")

plt.legend(handles=[dados_fra_plot, cdf_exp_plot, cdf_lognormal_plot])


for i in range(len(x)):
    plt.plot((x[i], x[i]), (fra[i], cdf_exponencial[i]), color="red", alpha=0.3)
    #plt.plot((x[i], x[i]), (fra[i], )
    plt.plot((x[i], x[i]), (fra[i], cdf_lognormal[i]), color="green", alpha=0.3)
```
